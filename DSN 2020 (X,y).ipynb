```python
#Importing neccessary libraries to use for the prediction
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
plt.rc("figure", figsize = (20,10))
```

```python
#Loading the dataset from my computer directory
train = pd.read_csv(r"C:\Users\lenovo\Desktop\DSN\2020 bookcamp competition\Train (1).csv")
test = pd.read_csv(r"C:\Users\lenovo\Desktop\DSN\2020 bookcamp competition\Test.csv")
sub = pd.read_csv(r"C:\Users\lenovo\Desktop\DSN\2020 bookcamp competition\SampleSubmission.csv")
```

```python
train.shape, test.shape, sub.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
train["form_field47"] = encoder.fit_transform(train["form_field47"])
test["form_field47"] = encoder.fit_transform(test["form_field47"])
train["default_status"] = encoder.fit_transform(train["default_status"])
```

```python
test["default_status"] = "test"
data = pd.concat([train, test])
```

```python
data["form_field47"].value_counts()
```

```python
data["default_status"].value_counts()
```

```python
data.info()
```

```python
data["form_field51"] = data["form_field8"] - data["form_field6"]
data["form_field52"] = data["form_field9"] - data["form_field8"]
data["form_field53"] = data["form_field9"] - data["form_field6"]
data["form_field54"] = data["form_field10"] + data["form_field6"]
data["form_field55"] = data["form_field10"] / data["form_field11"]
data["form_field56"] = data["form_field9"] - data["form_field8"] / data["form_field11"]
data["form_field57"] = data["form_field9"] - data["form_field6"] / data["form_field11"]
data["form_field58"] = data["form_field13"] + data["form_field10"]
data["form_field59"] = data["form_field16"] + (data["form_field17"] + data["form_field19"] /2)
data["form_field60"] = data["form_field18"] + (data["form_field20"]/2)
data["form_field61"] = data["form_field20"] * 0.75
data["form_field62"] = data["form_field19"] * 0.75
data["form_field63"] = data["form_field27"] - data["form_field25"]
data["form_field64"] = data["form_field14"] - data["form_field13"]
data["form_field65"] = data["form_field8"] - data["form_field6"] / data["form_field8"]
data["form_field66"] = data["form_field19"] - data["form_field17"]
data["form_field67"] = data["form_field20"] - data["form_field18"]
data["form_field68"] = data["form_field33"] + (data["form_field32"] / 30)
```

```python
data.isna().sum()
```

```python
data.isna().sum() * 100 / len(train)
```

```python
data.drop(["form_field31", "form_field40", "form_field41"], axis = 1, inplace =True)
```

```python
data.fillna(data.median(),inplace = True)
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.drop(["Applicant_ID"], axis = 1, inplace = True)
```

```python
cat_features = data.select_dtypes(exclude = np.number).columns.to_list()
num_features = data.select_dtypes(include = "number").columns.to_list()
print(cat_features)
print(num_features)
```

```python
data["default_status"].value_counts()
```

```python
train_df = data.loc[(data["default_status"] != "test")].reset_index(drop = True)
test_df = data.loc[(data["default_status"] == "test")].reset_index(drop = True)
```

```python
train_df.shape, test_df.shape
```

```python
X = train_df.drop(["default_status"], axis = 1)
y = train_df["default_status"].apply(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X.info()
```

```python
X[:] = np.nan_to_num(X)
```

```python
X_y = scaler.fit_transform(X,y)
```

```python
import xgboost as xgb
import lightgbm as lgbm
from catboost import CatBoostClassifier
```

```python
xgb = xgb.XGBClassifier(random_state = 1)
lgbm = lgbm.LGBMClassifier(random_state=1)
cat_boost = CatBoostClassifier()
```

```python
xgb.fit(X,y)
lgbm.fit(X,y)
cat_boost.fit(X,y)
```

```python
from sklearn.metrics import roc_auc_score
print("Xgboost AUC_ROC:{}".format (roc_auc_score(y, xgb.predict_proba(X)[:,1])))
print("Lightboost AUC_ROC:{}".format (roc_auc_score(y, lgbm.predict_proba(X)[:,1])))
print("CatBoostClassifier AUC_ROC:{}".format (roc_auc_score(y, cat_boost.predict_proba(X)[:,1])))
```

Baseline model
CatBoost Classifier = 90.7% 

```python
from sklearn.model_selection import cross_val_score
xgbcv = cross_val_score(xgb, X, y, cv = 5)
print(xgbcv.mean())
```

```python
lgbmcv = cross_val_score(lgbm, X, y, cv = 5)
print(lgbmcv.mean())
```

```python
from sklearn.model_selection import cross_val_score
catcv = cross_val_score(CatBoostClassifier(n_estimators =3000, eval_metric = "Accuracy", learning_rate = 0.1,
                              random_seed =1234), X, y, cv = 3, scoring = "roc_auc")
print(catcv.mean())
```

```python
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(estimators=[("xgb", xgb), ("lgbm", lgbm), ("cat_boost", cat_boost)], voting = "soft")
```

```python
votcv = cross_val_score(voting_clf, X, y, cv = 5, scoring = "roc_auc")
print(votcv)
print(votcv.mean())
```

```python
voting_clf.fit(X, y)
```

```python
test_x = test_df.drop("default_status", axis = 1)
```

```python
prediction = voting_clf.predict_proba(test_x)[:, 1]
```
