```python
def remove_collinear_features (x, threshold):
    y = x["default_status"]
    x = x.drop(columns = ["default_status"])
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) -1)
    drop_cols =[]
    for i in iters:
        for j in range(i):
            item = corr_matrix.iloc[j:(j+1), (i+1): (i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)
            if val>= threshold:
                drop_cols.append(col.values[0])
                drops = set(drop_cols)
                x = x.drop (columns = drops)
                x["default_status"] = y
                return x
```

```python
train_df = data.loc[(data["default_status"] != "test")].reset_index(drop = True)
test_df = data.loc[(data["default_status"] == "test")].reset_index(drop = True)
```

```python
X = train_df.drop(["default_status"], axis = 1)
y = train_df["default_status"].apply(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, stratify = y, random_state = 42)
```

```python
X.info()
```

```python
X_train[:] = np.nan_to_num(X_train)
```

```python
X_test[:] = np.nan_to_num(X_test)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
import xgboost as xgb
import lightgbm as lgbm
from catboost import CatBoostClassifier
```

```python
xgb = xgb.XGBClassifier(random_state = 1)
lgbm = lgbm.LGBMClassifier(random_state=1)
cat_boost = CatBoostClassifier()
```

```python
X.replace([np.inf, -np.inf], np.nan, inplace = True)
X.fillna(X.median(), inplace = True)
```

```python
# training algorithm
xgb.fit(X_train, y_train)
lgbm.fit(X_train, y_train)
cat_boost.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
print("Xgboost AUC_ROC:{}".format (roc_auc_score(y_test, xgb.predict_proba(X_test)[:,1])))
print("Lightboost AUC_ROC:{}".format (roc_auc_score(y_test, lgbm.predict_proba(X_test)[:,1])))
print("CatBoostClassifier AUC_ROC:{}".format (roc_auc_score(y_test, cat_boost.predict_proba(X_test)[:,1])))
```

```python
from catboost import CatBoostClassifier
error_list = []
test_pred = []
from sklearn.model_selection import StratifiedKFold
fold = StratifiedKFold(n_splits=2)
for train_index, test_index in fold.split(X,y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model = CatBoostClassifier(n_estimators =3000, eval_metric = "Accuracy", learning_rate = 0.1,
                              random_seed =1234, use_best_model = True)
    model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], early_stopping_rounds = 300,
             verbose = 100,)
    preds = model.predict(X_test)
    print("err:", roc_auc_score(y_test, preds))
    error_list.append(roc_auc_score(y_test, preds))
    p2 = model.predict(test_df.drop("default_status", axis = 1))
    test_pred.append(p2)
    np.mean(error_list)
```

```python
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(estimators=[("xgb", xgb), ("lgbm", lgbm), ("cat_boost", cat_boost)], voting = "soft")
```

```python
from sklearn.model_selection import cross_val_score
votcv = cross_val_score(voting_clf, X_train, y_train, cv = 3, scoring = "roc_auc")
print(votcv)
print(votcv.mean())
```
